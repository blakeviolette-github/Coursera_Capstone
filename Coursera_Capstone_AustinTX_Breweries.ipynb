## It looks like there aren't any breweries within 2 miles of the center of this zip code. Let's expand the search to all zip codes.

```python
def getNearbyVenues(names, latitudes, longitudes, radius=3218):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# create a new dataframe for all Auston venues

aus_ven = getNearbyVenues(names=df_auszip['Zip'],
                                   latitudes=df_auszip['Latitude'],
                                   longitudes=df_auszip['Longitude']
                                  )
```

# Methodology - filter the foursquare venues for Breweries and cluster the zip codes by number of breweries. This will identify which zip codes have the most breweries and therefore which zip code I should find a hotel in. 

```python
# filter for breweries
df_brew = aus_ven[aus_ven.vencat == 'Brewery']
df_brew.head()
```

```python
print("It looks like we found", df_brew['venue'].count(), "breweries in Austin! That's a lot of potential clients!")
```

```python
# count the breweries by zip code

df_brewzip = df_brew.groupby('Zip').count()
df_brewzip
```

```python
df_brewzip_count = pd.DataFrame(df_brewzip.vencat)
df_brewzip_count
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import k-means from clustering stage
from sklearn.cluster import KMeans
```

```python
# cluster the breweries

# set number of clusters
kclusters = 5


# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_brewzip_count)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# create a new df which includes the cluster

# add clustering labels
df_brewzip_count.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df_brewzip_count.reset_index()
```

```python
df_brewzipcount = pd.DataFrame(df_brewzip_count.reset_index())
```

```python
df_brewzipcount.columns
```

```python
# merge cluster number with zip code lat long df
aus_merged = df_auszip

aus_merged = aus_merged.join(df_brewzipcount.set_index('Zip'), on='Zip')

aus_merged.head()
```

```python
aus_merged.dropna(subset=['Cluster Labels'], inplace = True)
```

```python
aus_merged.astype({'Cluster Labels': 'int64'})
```

```python
# visualize the clusters
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors

# create map
map_clusters = map_aus

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, zipcode, cluster in zip(aus_merged['Latitude'], aus_merged['Longitude'], aus_merged['Zip'], aus_merged['Cluster Labels']):
    label = folium.Popup(str(zipcode) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster)],
        fill=True,
        fill_color=rainbow[int(cluster)],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
